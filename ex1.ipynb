```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

## Exercise 1b

```python
mu = 1
sigma = 2  # variance = 4

yrange = np.arange(1, np.exp(2*sigma), 0.1)  # define a range to calculate the pdf of Y
pdfy = np.exp( -(np.log(yrange) - mu)**2 / (2*sigma**2) ) / (yrange*sigma*np.sqrt(2*np.pi))  # compute the PDF with the formula

fig, ax = plt.subplots(1, 3, figsize=(14, 4))
for idx_n, n in enumerate([100, 1_000, 10_000]):
    
    x = sigma*np.random.randn(n) + mu  # n samples from a Gaussian with mean 1 and variance 4
    x = x[np.abs(x) < 2*sigma]  # discard samples outside the 96% (2*sigma) CI for better visualization
    y = np.exp(x)  # take the exponential of the samples

    ax[idx_n].hist(y, bins=30, density=True, label='Histogram')
    ax[idx_n].plot(yrange, pdfy, label='PDF')
    ax[idx_n].legend()
    ax[idx_n].set_title(f'n={n}')

plt.savefig('results/ex1b.pdf', bbox_inches='tight', pad_inches=0)
plt.draw()

```

## Exercise 1d

```python
lamda1 = 1  # call the variables "lamda" because "lambda" is the keyword for lambda functions
lamda2 = 2

maxy = 10  # keep the values below a threshold for better plotting

yrange = np.arange(0.1, maxy, 0.1)  # define a range to calculate the pdf of Y
pdfy = (lamda1*lamda2)/(lamda1 - lamda2) * (np.exp(-lamda2*yrange) - np.exp(-lamda1*yrange))  # compute the PDF with the formula

fig, ax = plt.subplots(1, 3, figsize=(14, 4))
for idx_n, n in enumerate([100, 1_000, 10_000]):

    u1 = np.random.rand(n)  # n samples uniform in (0, 1)
    u2 = np.random.ranf(n)

    y = -np.log(u1)/lamda1 -np.log(u2)/lamda2

    ax[idx_n].hist(y, bins=30, density=True, label='Histogram')
    ax[idx_n].plot(yrange, pdfy, label='PDF')
    ax[idx_n].legend()
    ax[idx_n].set_ylim([0, 0.8]) 
    ax[idx_n].set_title(f'n={n}')   
    

plt.savefig('results/ex1d.pdf', bbox_inches='tight', pad_inches=0)
plt.draw()
```
